```python
#Plotting simple function
price=[70000,74000,75000,80000,84000,74000]
year=[2018,2019,2020,2021,2022,2023]
plt.plot(year,price)
plt.grid()
plt.xlabel("Years")
plt.ylabel("Price of iPhone 8")
```

```python
#from pandas dataframe
batsman=pd.read_csv('sharma-kohli.csv')
batsman.head()
```

```python
batsman
```

```python
plt.plot(batsman['index'],batsman['V Kohli'])
```

```python
#plotting multiple plots
plt.plot(batsman['index'],batsman['V Kohli'])
plt.plot(batsman['index'],batsman['RG Sharma'])
```

```python
#labels and title
plt.plot(batsman['index'],batsman['V Kohli'])
plt.plot(batsman['index'],batsman['RG Sharma'])
plt.title("Virat Kohli Vs Rohit Sharma")
plt.xlabel("Season")
plt.ylabel("Runs Scored")
```

```python
## colors(hex) and line(width and style) and marker(size)
plt.plot(batsman['index'],batsman['V Kohli'],color='black')
plt.plot(batsman['index'],batsman['RG Sharma'],color='green')
```

```python
plt.plot(batsman['index'],batsman['V Kohli'],color='black',linestyle='dashdot')
plt.plot(batsman['index'],batsman['RG Sharma'],color='green',linestyle='dotted')
#linestle values: 'dashed' , 'dotted' , 'solid'(default),'dashdot'
```

```python
plt.plot(batsman['index'],batsman['V Kohli'],color='black',linestyle='dashdot',marker='.',markersize=10)
plt.plot(batsman['index'],batsman['RG Sharma'],color='green',linestyle='dotted',marker='+')
#markervalues: ".","+","D",.......
#size of marker can be changed by : markersize='num_value'
```

```python
#legend(location)
plt.plot(batsman['index'],batsman['V Kohli'],color='black',label='Virat')
plt.plot(batsman['index'],batsman['RG Sharma'],color='green',label='Rohit')
plt.legend(loc='best')
#values of loc:"best",'upper right', 'bottom right',............
```

```python
#limiting axes
plt.plot(batsman['index'],batsman['V Kohli'],color='black')
plt.plot(batsman['index'],batsman['RG Sharma'],color='green')
plt.xlim(2010,2015)
plt.ylim(200,800)
```

```python
#grid()
plt.plot(batsman['index'],batsman['V Kohli'],color='black')
plt.plot(batsman['index'],batsman['RG Sharma'],color='green')
plt.grid()
```
